```python
import pandas as pd
import numpy as np
import ast
```

```python
movie=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movie[['genres','keywords','overview','title']]
```

```python
df=movies.merge(credits,on='title')
```

```python
df.dropna(inplace=True)
```

```python
import ast
```

```python
text=[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
```

```python
def convert (text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df
```

```python
def cast(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
df['cast']=df['cast'].apply(cast)
```

```python
df
```

```python
text=df['crew'][0]
```

```python
df['crew'][0]
```

```python
def crew(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
             L.append(i['name'])
    return L
```

```python
df['crew']=df['crew'].apply(crew)
```

```python
df
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['genres']=df['genres'].apply(collapse)
df['keywords']=df['keywords'].apply(collapse)
df['cast']=df['cast'].apply(collapse)
df['crew']=df['crew'].apply(collapse)
```

```python
df
```

```python
df['overview']=df['overview'].apply(lambda x : x.split())
```

```python
df
```

```python
df['tags']=df['genres']+df['keywords']+df['cast']+df['crew']+df['overview']
```

```python
new=df.drop(columns=['genres','keywords','overview','cast','crew'])
```

```python
new
```

```python
new['tags']=new['tags'].apply(lambda x :" ".join(x))
```

```python
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Red Tails')
```
